```python
import pandas as pd
import numpy as np
import os
from matplotlib import pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python

def get_close_price(symbol):
    data_dir = r"C:\Users\Britta\Desktop\Active Equity Management\data"
    path = os.path.join(data_dir, symbol+'.csv')
    df = pd.read_csv(path)[['Date', 'Adj Close']]
    df['Date'] = pd.to_datetime(df['Date'])
    df['Date'] = df['Date'].astype(str)
    df.columns = ['date', 'close']
    # plt.plot(df['Adj Close'])
    # plt.show()
    df['year'] = df['date'].apply(lambda x: int(x[0:4]))
    df['month'] = df['date'].apply(lambda x: int(x[5:7]))
    return df


def security_trade(type):
    if type == 'stock':
        symbol = '^GSPC'
        start_date = '1993-10-25'
        end_date = '2012-05-05'
        buy_month = 11
        sell_month = 5
    elif type == 'bond':
        symbol = 'VUSTX'
        start_date = '1993-4-25'
        end_date = '2012-11-05'
        buy_month = 5
        sell_month = 11
    else:
        print("wrong type")
        return
    df = get_close_price(symbol)
    df = df[(df.date > start_date) & (df.date < end_date)]
    pdatas = df.reset_index(drop=True)
    pdatas['position'] = 0
    pdatas['flag'] = 0
    pricein = []
    priceout = []

    for i in range(len(pdatas)-1):
        if (pdatas.loc[i, 'month'] == buy_month-1) & (pdatas.loc[i+1, 'month'] == buy_month):
            pdatas.loc[i, 'flag'] = 1
            pdatas.loc[i+1, 'position'] = 1
            date_in = pdatas.loc[i, 'date']
            price_in = pdatas.loc[i, 'close']
            pricein.append([date_in, price_in])

        elif (pdatas.loc[i, 'month'] == sell_month-1) & (pdatas.loc[i+1, 'month'] == sell_month):
            pdatas.loc[i, 'flag'] = -1
            pdatas.loc[i+1, 'position'] = 0
            date_out = pdatas.loc[i, 'date']
            price_out = pdatas.loc[i, 'close']
            priceout.append([date_out, price_out])

        else:
            pdatas.loc[i + 1, 'position'] = pdatas.loc[i, 'position']

    p1 = pd.DataFrame(pricein, columns=['buydate', 'buyprice'])
    p2 = pd.DataFrame(priceout, columns=['selldate', 'sellprice'])

    transaction = pd.concat([p1, p2], axis=1)

    pdatas = pdatas.reset_index(drop=True)
    pdatas['ret'] = pdatas.close.pct_change(1).fillna(0)
    pdatas['benchmark'] = pdatas.close / pdatas.close[0]
    pdatas = pdatas[['date', 'position', 'ret', 'benchmark']]

    return transaction, pdatas


def strategy():
    trans1 = security_trade('stock')[0]
    trans2 = security_trade('bond')[0]
    transaction = pd.concat([trans1, trans2])
    transaction = transaction.sort_values('buydate').reset_index(drop=True)

    df1 = security_trade('stock')[1]
    df1.columns = ['date', 'pos1', 'ret1', 'benchmark1']
    df2 = security_trade('bond')[1]
    df2.columns = ['date', 'pos2', 'ret2', 'benchmark2']
    df = pd.merge(df1, df2, on='date')
    df['position'] = df['pos1']+df['pos2']
    df['ret'] = df['ret1']*df['pos1']+df['ret2']*df['pos2']
    df['nav'] = (1 + df.ret * df.position).cumprod()
    pdatas = df[['date', 'position', 'ret', 'nav', 'benchmark1', 'benchmark2']]
    return transaction, pdatas


def performance(transaction, strategy):
    # strategy = pdatas.copy();
    N = 250.0

    # annual return
    annual_ret = strategy.loc[len(strategy) - 1, 'nav'] ** (N / len(strategy)) - 1

    # Sharpe Ratio
    sharpe = (strategy['ret'] * strategy['position']).mean() / (strategy['ret'] * strategy['position']).std() * np.sqrt(
        N)

    # Winning Rate
    winning_rate = ((transaction['sellprice'] - transaction['buyprice']) > 0).mean()

    # Drawdown
    drawdown = 1 - strategy['nav'] / strategy['nav'].cummax()
    max_drawdown = max(drawdown)

    # Max Loss
    max_loss = min(transaction['sellprice'] / transaction['buyprice'] - 1)

    # Trades
    daily_trades = float(len(transaction)) / float(len(strategy))
    print(len(transaction))
    print(len(strategy))
    print("daily_trades", daily_trades)
    monthly_trades = daily_trades * 20

    # Annual performance of the strategy
    strategy['year'] = strategy['date'].apply(lambda x: x[:4])
    nav_peryear = strategy['nav'].groupby(strategy['year']).last() / strategy['nav'].groupby(strategy['year']).first()
    benchmark_peryear = strategy['benchmark2'].groupby(strategy['year']).last() / strategy['benchmark2'].groupby(
        strategy['year']).first()
    excess_ret = nav_peryear - benchmark_peryear
    result_peryear = pd.concat([nav_peryear, benchmark_peryear, excess_ret], axis=1)
    result_peryear.columns = ['strategy_ret', 'benchmark_ret', 'excess_ret']
    result_peryear = result_peryear.T

    # Graph
    xtick = np.round(np.linspace(0, len(strategy) - 1, 7), 0)
    xticklabel = strategy['date'][xtick]
    plt.figure(figsize=(9, 4))
    ax1 = plt.axes()
    plt.plot(np.arange(len(strategy)), strategy.benchmark1, color='black', label='S&P500', linewidth=2)
    plt.plot(np.arange(len(strategy)), strategy.benchmark2, color='blue', label='Tbond', linewidth=2)
    plt.plot(np.arange(len(strategy)), strategy.nav, color='red', label='nav', linewidth=2)
    # plt.plot(np.arange(len(strategy)), strategy.nav / strategy.benchmark2, color='orange', label='RS', linewidth=2)
    plt.legend()
    ax1.set_xticks(xtick)
    ax1.set_xticklabels(xticklabel)
    plt.show()

    # Output
    print('------------------------------')
    print('SharpeRatio:{}'.format(round(sharpe, 2)))
    print('AnnualReturn:{}%'.format(round(annual_ret * 100, 2)))
    print('WinningRate:{}%'.format(round(winning_rate * 100, 2)))
    print('MaxDrawDown:{}%'.format(round(max_drawdown * 100, 2)))
    print('MaxLoss:{}%'.format(round(-max_loss * 100, 2)))
    print('MonthlyNumberofTrades:{}(buy&sell)'.format(round(monthly_trades, 2)))

    result = {'Sharpe': sharpe,
              'annual_ret': annual_ret,
              'winning_rate': winning_rate,
              'max_drawdown': max_drawdown,
              'max_loss': max_loss,
              'daily_trades': daily_trades
              }
    result = pd.DataFrame.from_dict(result, orient='index').T

    return result, result_peryear


def backtest():
    transaction, pdatas = strategy()
    stats, result_peryear = performance(transaction, pdatas)
    return stats, result_peryear
```
